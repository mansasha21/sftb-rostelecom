# catboost

```python
import numpy as np
import pandas as pd
import seaborn as sns
from catboost import CatBoostClassifier
from sklearn.preprocessing import MinMaxScaler
from sklearn.metrics import mean_absolute_error, r2_score, mean_squared_error, roc_auc_score
from sklearn.model_selection import train_test_split, StratifiedKFold
from sklearn.metrics import classification_report
from matplotlib import pyplot as plt
```

```python
train = pd.read_csv("./data/train.csv",sep=";",index_col=False)
test = pd.read_csv("./data/test.csv",sep=";",index_col=False)
target = train['label']
try_set = train
```

```python
def process_period(try_set):
    try_set['period'] = pd.to_datetime(try_set['period'])
    try_set['year'] = try_set['period'].dt.year
    try_set['month'] = try_set['period'].dt.month
    try_set['day'] = try_set['period'].dt.day
    drop_cols = ['period']
    try_set.drop(drop_cols, axis=1, inplace=True)
    return try_set
try_set = process_period(try_set)
test = process_period(test)
```

```python
# drop nan_city_name from try_set
def process_nan(try_set):
    try_set = try_set[~try_set.city_name.isnull()]
    try_set = try_set.dropna(axis=1, thresh=len(try_set) * 0.8)
    try_set = try_set.replace(np.nan, 0.0)
    return try_set
try_set = process_nan(try_set)
test = process_nan(test)
```

```python
print(test.columns)
print(try_set.columns)
```

```python
try_set.head()
```

```python
def mk_cat(try_set):
    try_set[['subject_type', 'subject_name', 'city_name']] = try_set[['subject_type', 'subject_name', 'city_name']].apply(lambda x: x.astype('category'))
    try_set.hex = try_set['hex'].apply(int, base=16)
    try_set.reset_index(drop=True, inplace=True)
    
    return try_set
try_set = mk_cat(try_set)
test = mk_cat(test)

target = try_set['label']
try_set = try_set.drop(['label'], axis=1)
```

```python
test.head()
```

```python
cat_features = try_set.select_dtypes(include=['category']).columns.to_list()

skfg = StratifiedKFold(n_splits=3, random_state=142, shuffle=True)
fold_id = 0
models = []
for train_ids, val_ids in skfg.split(try_set, target):
    classifier = CatBoostClassifier(random_state=142, iterations=4000,
                                    verbose=250, use_best_model=True, early_stopping_rounds=500,
                                    eval_metric='NormalizedGini', loss_function = 'Logloss')
                                    #class_weights=[1.55839206, 0.59422713, 1.48048405])    
    classifier.fit(try_set.loc[train_ids], target[train_ids], cat_features=cat_features,
                   eval_set=(try_set.loc[val_ids], target[val_ids]))
    models.append(classifier)
```

```python
test.shape
```

```python
y_pred = models[0].predict(test.drop(['label'], axis=1))
print(classification_report(test['label'], y_pred))
from sklearn import metrics

def draw_roc_auc(y_true, y_pred):
    print(roc_auc_score(y_true, y_pred))
    fpr, tpr, thresholds = metrics.roc_curve(y_true, y_pred)
    plt.figure(figsize=(10,6))
    plt.plot([0, 1], [0, 1], color='navy', lw=2, linestyle='--')
    plt.plot(fpr,tpr)
    plt.show()


draw_roc_auc(test['label'], y_pred)
```

```python
def plot_feature_importance(importance,names,model_type):
    
    #Create arrays from feature importance and feature names
    feature_importance = np.array(importance)
    feature_names = np.array(names)
    
    #Create a DataFrame using a Dictionary
    data={'feature_names':feature_names,'feature_importance':feature_importance}
    fi_df = pd.DataFrame(data)
    
    #Sort the DataFrame in order decreasing feature importance
    fi_df.sort_values(by=['feature_importance'], ascending=False,inplace=True)
    
    #Define size of bar plot
    plt.figure(figsize=(10,8))
    #Plot Searborn bar chart
    sns.barplot(x=fi_df['feature_importance'], y=fi_df['feature_names'])
    #Add chart labels
    plt.title(model_type + 'FEATURE IMPORTANCE')
    plt.xlabel('FEATURE IMPORTANCE')
    plt.ylabel('FEATURE NAMES')
plot_feature_importance(models[0].feature_importances_,try_set.columns, 'CatBoost')
```
